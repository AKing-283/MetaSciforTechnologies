```python
import requests

url = 'https://api.quotable.io/random'
response =  requests.get(url)
data=response.json()
if response.status_code==200:
  author = data.get('author')
  quote = data.get('content')
  print("-----Welcome to the quotes engine-------")
  print(f'"{quote}-{author}"')
```

```python
import random
import string

def password_gen(length,use_char=True,use_num=True,use_special=True):
  use_char = bool(input("Do you want to include characters? "))
  use_num = bool(input("Do you want to include numbers?"))
  use_special = bool(input("Do you want to include special characters?"))
  length = int(input("Enter the length of password you want: "))
  password =""
  for i in range(length):
    if use_char == True and use_num==True and use_special==True:
      characters = string.ascii_letters
      numbers = string.digits if use_num==True else ""
      special_chars = string.punctuation if use_special==True else ""
      length = max(length,1)
      gen_pwd = characters+numbers+special_chars
      password = ''.join(random.choice(gen_pwd)for i in range(length))
      return password

pwd1 = password_gen(7,True,True,True)
print("The password generated by program is: ",pwd1)
```
